```python
 # Initial imports
import os
import requests
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi

%matplotlib inline
```

```python
load_dotenv()
```

```python
 # Set Alpaca API key and secret
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")
```

```python
print(alpaca_api_key)
print(alpaca_secret_key)
```

```python
 # Verify that Alpaca key and secret were correctly loaded
print(f"Alpaca Key type: {type(alpaca_api_key)}")
print(f"Alpaca Secret Key type: {type(alpaca_secret_key)}")
```

```python
alpaca=tradeapi.REST(alpaca_api_key, alpaca_secret_key, api_version="v2")
```

```python
from alpaca_trade_api.rest import REST, TimeFrame
tickers = ["FB", "TWTR"]
```

```python
today = pd.Timestamp("2020-07-14", tz="America/New_York").isoformat()
today1 = pd.Timestamp("2021-07-14", tz="America/New_York").isoformat()
```

```python
 # Get current closing prices for FB and TWTR
yearly = alpaca.get_bars(
    tickers,
    TimeFrame.Day,
    start = today,
    end = today1
).df

# Display sample data
yearly
```

```python
yearly.index=yearly.index.date
```

```python
yearly = yearly[["close", "symbol"]]
yearly
```

```python
tables = pd.pivot_table(yearly, values='close', index=yearly.index, columns=["symbol"])
tables
```
